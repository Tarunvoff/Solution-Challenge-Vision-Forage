```python
from pymongo import MongoClient
from datetime import datetime
import os
from typing import Dict, Any, List, Optional
from google.colab import userdata

class MongoDBManager:
    def __init__(self, uri: str = None):
        mongo_uri = os.getenv("MONGODB_URI", userdata.get("Mango"))
        self.client = MongoClient(mongo_uri)  # ✅ Now self.client is defined

        self.db = self.client["chatbotDB"]  # ✅ No error now

        # Initialize collections (if needed)
        self.messages_collection = self.db["messages"]
        self.users_collection = self.db["users"]
        # Initialize collections
        self.chats = self.db["chats"]
        self.feedback = self.db["feedback"]
        self.users = self.db["users"]
        self.sessions = self.db["sessions"]

    def store_chat(self, user_message: str, bot_response: str,
                  intent: str, language: str, user_id: Optional[str] = None) -> str:
        """
        Store a chat interaction in the database
        Returns the chat_id
        """
        chat_data = {
            "user_message": user_message,
            "bot_response": bot_response,
            "intent": intent,
            "language": language,
            "user_id": user_id,
            "timestamp": datetime.utcnow()
        }
        result = self.chats.insert_one(chat_data)
        return str(result.inserted_id)

    def store_feedback(self, chat_id: str, rating: str,
                      reason: Optional[str] = None, user_id: Optional[str] = None) -> bool:
        """
        Store user feedback for a chat interaction
        Returns True if successful
        """
        feedback_data = {
            "chat_id": chat_id,
            "rating": rating,
            "reason": reason,
            "user_id": user_id,
            "timestamp": datetime.utcnow()
        }
        try:
            self.feedback.insert_one(feedback_data)
            return True
        except Exception as e:
            print(f"Error storing feedback: {str(e)}")
            return False

    def store_user(self, user_data: Dict[str, Any]) -> str:
        """
        Store or update user information
        Returns the user_id
        """
        user_data["last_updated"] = datetime.utcnow()
        result = self.users.update_one(
            {"user_id": user_data.get("user_id")},
            {"$set": user_data},
            upsert=True
        )
        return str(result.upserted_id) if result.upserted_id else user_data.get("user_id")

    def store_session(self, session_data: Dict[str, Any]) -> str:
        """
        Store a new chat session
        Returns the session_id
        """
        session_data["created_at"] = datetime.utcnow()
        result = self.sessions.insert_one(session_data)
        return str(result.inserted_id)

    def get_chat_history(self, user_id: str, limit: int = 10) -> List[Dict[str, Any]]:
        """
        Retrieve chat history for a specific user
        """
        return list(self.chats.find(
            {"user_id": user_id}
        ).sort("timestamp", -1).limit(limit))

    def get_user_feedback(self, user_id: str, limit: int = 10) -> List[Dict[str, Any]]:
        """
        Retrieve feedback history for a specific user
        """
        return list(self.feedback.find(
            {"user_id": user_id}
        ).sort("timestamp", -1).limit(limit))

    def close(self):
        """
        Close the MongoDB connection
        """
        self.client.close()

# Example usage
if __name__ == "__main__":
    # Initialize MongoDB manager
    mongo_manager = MongoDBManager()

    try:
        # Example: Store a chat
        chat_id = mongo_manager.store_chat(
            user_message="Hello!",
            bot_response="Hi there!",
            intent="greeting",
            language="en",
            user_id="user123"
        )
        print(f"Stored chat with ID: {chat_id}")

        # Example: Store feedback
        success = mongo_manager.store_feedback(
            chat_id=chat_id,
            rating="👍",
            reason="Good response",
            user_id="user123"
        )
        print(f"Stored feedback: {success}")

        # Example: Get chat history
        chat_history = mongo_manager.get_chat_history("user123")
        print("Chat history:", chat_history)

    finally:
        # Always close the connection
        mongo_manager.close()
```

```python
from flask import Flask, request, jsonify
import google.generativeai as genai
import json
import random
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer
from flask_cors import CORS
import os
from datetime import datetime
from deep_translator import GoogleTranslator
from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from google.colab import userdata

# ✅ Load Gemini API key securely
API_KEY = userdata.get("GOOGLE_API_KEY_2")
if not API_KEY:
    raise ValueError("API Key not found. Set GEMINI_API_KEY in your environment.")

genai.configure(api_key=API_KEY)

# ✅ Load intents from JSON
with open("./intents.json", "r", encoding="utf-8") as file:
    intents_data = json.load(file)

# ✅ Initialize Gemini for intent detection
gemini_model = genai.GenerativeModel("gemini-1.5-flash-latest")

# ✅ Load HELPING AI Model from Hugging Face
HELPING_AI_MODEL = "OEvortex/HelpingAI-3B"
tokenizer = AutoTokenizer.from_pretrained(HELPING_AI_MODEL)
helping_model = AutoModelForCausalLM.from_pretrained(HELPING_AI_MODEL)

# ✅ Initialize MongoDB manager
mongo_manager = MongoDBManager()

# ✅ Initialize Sentence Transformer Model
embedding_model = SentenceTransformer("all-MiniLM-L6-v2")

# ✅ Prepare FAISS Index
vector_dim = 384  # Embedding size for MiniLM
index = faiss.IndexFlatL2(vector_dim)
intent_tags = []  # List to store intent tags

# ✅ Precompute and store embeddings for intents
for intent in intents_data["intents"]:
    embedding = embedding_model.encode(intent["patterns"]).mean(axis=0)  # Average embedding
    index.add(np.array([embedding], dtype=np.float32))
    intent_tags.append(intent["tag"])

# ✅ Initialize Flask App
app = Flask(__name__)
CORS(app)

def detect_language(text):
    return GoogleTranslator().detect(text)

def translate_to_english(text, src_lang):
    if src_lang != "en":
        return GoogleTranslator(source=src_lang, target="en").translate(text)
    return text

def get_similar_intent(user_input):
    try:
        # Handle empty input
        if not user_input or not user_input.strip():
            return "unknown_intent"

        # Encode user input and handle potential encoding errors
        try:
            user_embedding = embedding_model.encode(user_input)
        except Exception as e:
            print(f"Error encoding user input: {str(e)}")
            return "unknown_intent"

        # Ensure embedding is the correct shape/type
        if not isinstance(user_embedding, np.ndarray):
            user_embedding = np.array(user_embedding)
        user_embedding = user_embedding.reshape(1, -1).astype(np.float32)

        # Search for similar intents with error handling
        try:
            D, I = index.search(user_embedding, k=1)
        except Exception as e:
            print(f"Error searching FAISS index: {str(e)}")
            return "unknown_intent"

        # Validate search results
        if len(I) == 0 or len(I[0]) == 0:
            return use_gemini_fallback(user_input)

        index_value = I[0][0]

        # Check confidence threshold
        if D[0][0] > 10.0:  # High distance means low confidence
            return use_gemini_fallback(user_input)

        # Validate index bounds
        if 0 <= index_value < len(intent_tags):
            return intent_tags[index_value]
        return use_gemini_fallback(user_input)

    except Exception as e:
        print(f"Unexpected error in get_similar_intent: {str(e)}")
        return use_gemini_fallback(user_input)

def use_gemini_fallback(user_input: str) -> str:
    """
    Use Gemini to detect intent when FAISS search fails or has low confidence
    """
    try:
        prompt = f"""
        Analyze the following user message and determine the most appropriate intent category.
        The intent should be a single word or short phrase in snake_case format (e.g., greeting, farewell, help_request).
        If you're unsure, return "general_query".

        User message: "{user_input}"

        Intent:"""

        response = gemini_model.generate_content(prompt)
        detected_intent = response.text.strip().lower()

        # Clean up the response and ensure it's in snake_case
        detected_intent = detected_intent.replace(" ", "_")
        detected_intent = ''.join(c for c in detected_intent if c.isalnum() or c == '_')

        # Validate the response
        if not detected_intent:
            return "general_query"

        return detected_intent

    except Exception as e:
        print(f"Error in Gemini fallback: {str(e)}")
        return "general_query"

def get_intent_response(intent_tag):
    for intent in intents_data["intents"]:
        if intent["tag"] == intent_tag:
            return random.choice(intent["responses"])
    return "I'm not sure how to respond to that."

def generate_helping_ai_response(user_input, detected_intent, base_response):
    prompt = f"""
    User Input: {user_input}
    Detected Intent: {detected_intent}
    Base Response: {base_response}
    AI Response:
    """
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=512, padding=True)
    with torch.no_grad():
        output = helping_model.generate(**inputs, max_new_tokens=60, pad_token_id=tokenizer.eos_token_id)
    response = tokenizer.decode(output[0], skip_special_tokens=True).strip()
    return response.split("AI Response:")[-1].strip() if "AI Response:" in response else response

@app.route('/chat', methods=['POST'])
def chat():
    try:
        data = request.get_json()
        user_message = data.get("message", "")
        user_id = data.get("user_id", "anonymous")
        session_id = data.get("session_id")

        # Detect language and translate if needed
        detected_lang = detect_language(user_message)
        translated_message = translate_to_english(user_message, detected_lang)

        # Get intent and generate response
        detected_intent = get_similar_intent(translated_message)
        base_response = get_intent_response(detected_intent)
        final_response = generate_helping_ai_response(translated_message, detected_intent, base_response)
        final_response = GoogleTranslator(source="en", target=detected_lang).translate(final_response)

        # Store chat in MongoDB
        chat_id = mongo_manager.store_chat(
            user_message=user_message,
            bot_response=final_response,
            intent=detected_intent,
            language=detected_lang,
            user_id=user_id
        )

        # If this is a new session, create one
        if not session_id:
            session_data = {
                "user_id": user_id,
                "start_time": datetime.utcnow(),
                "language": detected_lang
            }
            session_id = mongo_manager.store_session(session_data)

        return jsonify({
            "response": final_response,
            "chat_id": chat_id,
            "session_id": session_id,
            "intent": detected_intent
        })

    except Exception as e:
        print(f"Error in chat endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500

@app.route('/feedback', methods=['POST'])
def feedback():
    try:
        data = request.get_json()
        chat_id = data.get("chat_id")
        rating = data.get("rating")  # Expecting "👍" or "👎"
        reason = data.get("reason", "")
        user_id = data.get("user_id", "anonymous")

        # Store feedback in MongoDB
        success = mongo_manager.store_feedback(
            chat_id=chat_id,
            rating=rating,
            reason=reason,
            user_id=user_id
        )

        return jsonify({"message": "Feedback recorded successfully"})

    except Exception as e:
        print(f"Error in feedback endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500

@app.route('/chat_history', methods=['GET'])
def get_chat_history():
    try:
        user_id = request.args.get("user_id")
        limit = int(request.args.get("limit", 10))

        if not user_id:
            return jsonify({"error": "user_id is required"}), 400

        chat_history = mongo_manager.get_chat_history(user_id, limit)
        return jsonify({"chat_history": chat_history})

    except Exception as e:
        print(f"Error in chat_history endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500

@app.route('/user_feedback', methods=['GET'])
def get_user_feedback():
    try:
        user_id = request.args.get("user_id")
        limit = int(request.args.get("limit", 10))

        if not user_id:
            return jsonify({"error": "user_id is required"}), 400

        feedback_history = mongo_manager.get_user_feedback(user_id, limit)
        return jsonify({"feedback_history": feedback_history})

    except Exception as e:
        print(f"Error in user_feedback endpoint: {str(e)}")
        return jsonify({"error": str(e)}), 500

@app.teardown_appcontext
def close_db(error):
    """Close MongoDB connection when the app context is torn down"""
    mongo_manager.close()

if __name__ == '__main__':
    app.run(debug=True)
```
